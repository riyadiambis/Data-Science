```python
sesuatu = pd.read_csv('Human_Resources.csv')
sesuatu
```

```python
#Pilih baris tertentu yang memenuhi kriteria tertentu

karyawan_loyal=sesuatu[(sesuatu['YearsAtCompany']>=30)]
karyawan_loyal
```

```python
karyawan_loyal['YearsAtCompany']
```

```python
loyal_terfilter=karyawan_loyal[(karyawan_loyal['YearsAtCompany']>=33)&(karyawan_loyal['YearsAtCompany']<=35)]
loyal_terfilter['YearsAtCompany']
```

```python
kriteria1= sesuatu['YearsAtCompany'] >= 25
kriteria2= sesuatu['Department'] == 'Research & Development'
karyawan_mantap=sesuatu[kriteria1 & kriteria2]

karyawan_mantap
```
